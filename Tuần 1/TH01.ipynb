```python
import pandas as pd
import numpy as np
import os

pd.set_option("display.max_columns", None)
```

#### Tải dữ liệu xét tuyển đại học

```python
df = pd.read_csv("../DuLieu/dulieuxettuyendaihoc.csv", index_col = "STT")
df.head(10)
```

#### Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0.

```python
df["DT"] = df["DT"].fillna(0)
df
```

#### Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean. 

```python
df["T1"] = df["T1"].fillna(df["T1"].mean())
df
```

#### Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.

```python
cols_score = [
    'T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]

for col in cols_score:
    if col in df.columns: # Kiểm tra xem cột có trong dữ liệu không
        # Tính giá trị trung bình của cột
        mean_val = df[col].mean()
        
        # Điền giá trị trung bình vào các ô bị thiếu (fillna)
        df[col] = df[col].fillna(mean_val)

df.head(10)
```

#### Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và 12. 

```python
cols_score = ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
              'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
              'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6']

for col in cols_score:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())

# 3. Tính TBM1, TBM2, TBM3 theo công thức: (Toán*2 + Văn*2 + Các môn khác) / 10
# TBM1: Lớp 10 (đuôi 1)
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + 
              df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10

# TBM2: Lớp 11 (đuôi 2)
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + 
              df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10

# TBM3: Lớp 12 (đuôi 6 - theo mô tả dữ liệu)
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + 
              df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

# 4. Làm tròn kết quả (2 chữ số thập phân) và hiển thị
df[['TBM1', 'TBM2', 'TBM3']] = df[['TBM1', 'TBM2', 'TBM3']].round(2)

print("5 dòng đầu tiên của các cột TBM vừa tạo:")
print(df[['TBM1', 'TBM2', 'TBM3']].head())
```

#### Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:

```python
# 1. Định nghĩa hàm xếp loại theo quy tắc đề bài
def xep_loai(score):
    if score < 5.0:
        return 'Y'   # Yếu
    elif score < 6.5:
        return 'TB'  # Trung bình
    elif score < 8.0:
        return 'K'   # Khá
    elif score < 9.0:
        return 'G'   # Giỏi
    else:
        return 'XS'  # Xuất sắc

# 2. Áp dụng hàm này cho từng cột TBM1, TBM2, TBM3
# Cậu có thể dùng vòng lặp để code gọn hơn
for i in ['1', '2', '3']:
    col_tbm = f'TBM{i}'  # Tên cột điểm (ví dụ: TBM1)
    col_xl = f'XL{i}'    # Tên cột xếp loại mới (ví dụ: XL1)
    
    # Dùng hàm .apply() để chạy hàm xep_loai cho từng dòng dữ liệu
    df[col_xl] = df[col_tbm].apply(xep_loai)

# 3. Xem kết quả
print(df[['TBM1', 'XL1', 'TBM2', 'XL2', 'TBM3', 'XL3']].head())
```

#### Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
# 1. Tạo các biến US_TBM dựa trên TBM đã tính
# Công thức: Điểm hệ 4 = (Điểm hệ 10 / 10) * 4
for i in ['1', '2', '3']:
    tbm_col = f'TBM{i}'      # Tên cột điểm hệ 10 (ví dụ: TBM1)
    us_tbm_col = f'US_TBM{i}' # Tên cột điểm hệ 4 mới (ví dụ: US_TBM1)
    
    # Áp dụng công thức Min-Max Normalization
    df[us_tbm_col] = (df[tbm_col] / 10) * 4

# 2. Làm tròn 3 chữ số thập phân (GPA Mỹ thường lấy 2-3 số lẻ)
df[['US_TBM1', 'US_TBM2', 'US_TBM3']] = df[['US_TBM1', 'US_TBM2', 'US_TBM3']].round(3)

# 3. Hiển thị kết quả so sánh
print("So sánh điểm hệ 10 và hệ 4:")
print(df[['TBM1', 'US_TBM1', 'TBM2', 'US_TBM2', 'TBM3', 'US_TBM3']].head())
```

#### Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3

```python
# 1. Định nghĩa hàm xử lý logic xét tuyển
def xet_tuyen(row):
    # Lấy giá trị từng cột trên dòng hiện tại
    dh1 = row['DH1']
    dh2 = row['DH2']
    dh3 = row['DH3']
    kt = row['KT']
    
    # Tính điểm trung bình (dtb) theo khối thi
    if kt in ['A', 'A1']:
        dtb = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        dtb = (dh1 + dh2*2 + dh3) / 4
    else:
        dtb = (dh1 + dh2 + dh3) / 3
    
    # Trả về kết quả: 1 (Đậu) hoặc 0 (Rớt)
    if dtb >= 5.0:
        return 1
    else:
        return 0

# 2. Áp dụng hàm xet_tuyen cho toàn bộ DataFrame
# axis=1 nghĩa là áp dụng theo từng dòng (row)
df['KQXT'] = df.apply(xet_tuyen, axis=1)

# 3. Hiển thị kết quả kiểm tra
print("10 dòng đầu tiên của kết quả xét tuyển:")
print(df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10))
```

#### Lưu trữ dữ liệu xuống ổ đĩa thành file processed_dulieuxettuyendaihoc.csv

```python
# Q6: Xử lý dữ liệu thiếu
cols_score = ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
              'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
              'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
              'DH1', 'DH2', 'DH3']
for col in cols_score:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())

# Q7: Tính TBM
for i in ['1', '2', '6']:
    tbm_col = "TBM" + ("3" if i == "6" else i)
    df[tbm_col] = (df["T"+i]*2 + df["L"+i] + df["H"+i] + df["S"+i] + 
                   df["V"+i]*2 + df["X"+i] + df["D"+i] + df["N"+i]) / 10
df[['TBM1', 'TBM2', 'TBM3']] = df[['TBM1', 'TBM2', 'TBM3']].round(2)

# Q8: Xếp loại (XL)
def xep_loai(score):
    if score < 5.0: return 'Y'
    elif score < 6.5: return 'TB'
    elif score < 8.0: return 'K'
    elif score < 9.0: return 'G'
    else: return 'XS'
for i in ['1', '2', '3']:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)

# Q9: Chuyển đổi thang điểm Mỹ (US_TBM)
for i in ['1', '2', '3']:
    df[f'US_TBM{i}'] = (df[f'TBM{i}'] / 10) * 4
df[['US_TBM1', 'US_TBM2', 'US_TBM3']] = df[['US_TBM1', 'US_TBM2', 'US_TBM3']].round(3)

# Q10: Kết quả xét tuyển (KQXT)
def xet_tuyen(row):
    dh1, dh2, dh3, kt = row['DH1'], row['DH2'], row['DH3'], row['KT']
    if kt in ['A', 'A1']:
        dtb = (dh1*2 + dh2 + dh3) / 4
    elif kt == 'B':
        dtb = (dh1 + dh2*2 + dh3) / 4
    else:
        dtb = (dh1 + dh2 + dh3) / 3
    return 1 if dtb >= 5.0 else 0
df['KQXT'] = df.apply(xet_tuyen, axis=1)

# Lưu dataframe đã xử lý ra file CSV
filename = 'processed_dulieuxettuyendaihoc.csv'
df.to_csv(filename, index=False)

print(f" Đã lưu thành công file: {filename}")
```
